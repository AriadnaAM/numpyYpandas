##### a. Extraer la información del archivo.

```python
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/gdrive',force_remount=True)
```

```python
country_vaccinations=pd.read_csv('/content/gdrive/MyDrive/country_vaccinations.csv')
```

```python
country_vaccinations=pd.DataFrame(country_vaccinations)
country_vaccinations.head(5)
```

##### b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64.

```python
#estructura y tipos de datos
country_vaccinations.info()
```

```python
#columna "date" en formato de fecha datetime64
country_vaccinations['date'] = pd.to_datetime(country_vaccinations['date'])
country_vaccinations['date'].head(5)
```

```python
# formato de datos con ajuste en fecha
country_vaccinations.info()
```

##### c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
#Seleccion de columnas de compañias y vacunas aplicadas
vaccines=country_vaccinations[["vaccines","total_vaccinations"]]
vaccines.head(5)
```

```python
#Resumen de aplicación de vacunas por compañías
resumen = round(vaccines.groupby('vaccines')['total_vaccinations'].sum(),2)
resumen=pd.DataFrame(resumen)
resumen.head()
```

#####d. Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
total = round(vaccines['total_vaccinations'].sum(),2)
print('La cantidad de vacunas aplicadas en todo el mundo es:',total)
```

#####e. Calcular el promedio de vacunas aplicadas por país.

```python
#Seleccion de columnas de paises y vacunas aplicadas
countries=country_vaccinations[["country","total_vaccinations"]]
```

```python
#Resumen de aplicación de vacunas por país
resumen_country = round(countries.groupby('country')['total_vaccinations'].mean(),2)
resumen_country=pd.DataFrame(resumen_country)
resumen_country.head()
```

#####f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.

```python
vaccinations29 = country_vaccinations[(country_vaccinations['date'] == '2021-01-29')][['country', 'date', 'total_vaccinations']]
total29 = round(vaccinations29['total_vaccinations'].sum(),2)
print('El total de vacunas aplicadas el 29/01/2021 en el mundo fue:',total29)
```

#####g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.**  

```python
conDiferencias=pd.DataFrame(country_vaccinations.copy())
conDiferencias['Diferencias']=conDiferencias['daily_vaccinations']-conDiferencias['daily_vaccinations_raw']
conDiferencias.head()
```

##### h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
periodo=country_vaccinations['date'].max()-country_vaccinations['date'].min()
print('El periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua es:', periodo)
```

#####i.Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
conCantidad=pd.DataFrame(country_vaccinations.copy())
conCantidad['canVac']=conCantidad['daily_vaccinations']
conCantidad.head()
```

#####j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = country_vaccinations[(country_vaccinations['date'] <= '2020-12-20')]
antes20.head()
```

#####k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer=pd.DataFrame(country_vaccinations.copy())
pfizer= pfizer[pfizer['vaccines'].str.contains('Pfizer')]
pfizer.head()
```

#####l. Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente.

```python
with pd.ExcelWriter('/content/gdrive/MyDrive/resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name="conDiferencias")
    conCantidad.to_excel(writer, sheet_name="conCantidad")
    antes20.to_excel(writer, sheet_name="antes20")
    pfizer.to_excel(writer, sheet_name="pfizer")
```
